```python
# First import packages

import pandas as pd
import numpy as np
import pdb
import seaborn as sns
from matplotlib import pyplot as plt
import os
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
#Set the color for each projection population

colors = sns.color_palette('muted', 10)

PFC_color = "#FFDB6E"
NAc_color = "#BF43CC"
LH_color = "#87FFE1"
```

```python
#Load the .csv file for coronal CTB data
df = pd.read_csv("coronal_CTBLabelled_dataset.csv")
```

```python
#Filter dataframe by projection type
df_pfc = df[df.Projection == 'PFC']
df_nac = df[df.Projection == 'NAc']
df_lh = df[df.Projection == 'LH']
```

```python
# Plotting the data in 3D

PFC_color = "#FFDB6E"
NAc_color = "#BF43CC"
LH_color = "#87FFE1"

fig = go.Figure(data=[go.Scatter3d(
        x=df_pfc["AP"],
        y=df_pfc["ML"],
        z=df_pfc["DV"],
        mode= "markers", 
        marker = dict(size = 2, opacity = 0.5, color = PFC_color),
        name="PFC"),

    go.Scatter3d(
        x=df_nac["AP"],
        y=df_nac["ML"],
        z=df_nac["DV"],
        mode= "markers", 
        marker = dict(size = 2, opacity = 0.5, color = NAc_color),
        name="NAc"),

    go.Scatter3d(
        x=df_lh["AP"],
        y=df_lh["ML"],
        z=df_lh["DV"],
        mode= "markers", 
        marker = dict(size = 2, opacity = 0.5, color = LH_color),
        name="LH"),
                      
                     ])
fig.update_layout(
    title="Projection",
    scene=dict(
        xaxis=dict(title='AP'),
        yaxis=dict(title='ML'),
        zaxis=dict(title='DV')
    ),
    width=1200,
    height=800,
    autosize=False
)

fig.update_scenes(aspectmode="data")

fig.show()
```
