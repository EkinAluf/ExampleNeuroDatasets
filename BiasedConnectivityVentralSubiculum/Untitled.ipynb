```python
# Load the .csv files

lh = pd.read_csv("LH_TRIO_cellcounts.csv")
nac = pd.read_csv("NAc_TRIO_cellcounts.csv")
pfc = pd.read_csv("PFC_TRIO_cellcounts.csv")
```

```python
def compute_ecdf(df, regions, axis="AP"):
    """
    df: dataframe (lh, nac, pfc)
    regions: list of acronyms (e.g. ["RE"] or ["MS", "NDB"])
    axis: which coordinate to use (default AP)
    """
    
    # filter regions
    subset = df[df["acronym"].isin(regions)]
    
    # extract values
    values = subset[axis].dropna()
    
    # sort
    x = np.sort(values)
    
    # cumulative proportion
    y = np.arange(1, len(x) + 1) / len(x)
    
    return x, y, values.min(), values.max()
```

```python
x_lh_re, y_lh_re, min_lh_re, max_lh_re = compute_ecdf(lh, ["RE"])
x_nac_re, y_nac_re, min_nac_re, max_nac_re = compute_ecdf(nac, ["RE"])
x_pfc_re, y_pfc_re, min_pfc_re, max_pfc_re = compute_ecdf(pfc, ["RE"])

x_lh_pa, y_lh_pa, min_lh_pa, max_lh_pa = compute_ecdf(lh, ["PA"])
x_nac_pa, y_nac_pa, min_nac_pa, max_nac_pa = compute_ecdf(nac, ["PA"])
x_pfc_pa, y_pfc_pa, min_pfc_pa, max_pfc_pa = compute_ecdf(pfc, ["PA"])

x_lh_mpo, y_lh_mpo, min_lh_mpo, max_lh_mpo = compute_ecdf(lh, ["MPO"])
x_nac_mpo, y_nac_mpo, min_nac_mpo, max_nac_mpo = compute_ecdf(nac, ["MPO"])
x_pfc_mpo, y_pfc_mpo, min_pfc_mpo, max_pfc_mpo = compute_ecdf(pfc, ["MPO"])

x_lh_pvt, y_lh_pvt, min_lh_pvt, max_lh_pvt = compute_ecdf(lh, ["PVT"])
x_nac_pvt, y_nac_pvt, min_nac_pvt, max_nac_pvt = compute_ecdf(nac, ["PVT"])
x_pfc_pvt, y_pfc_pvt, min_pfc_pvt, max_pfc_pvt = compute_ecdf(pfc, ["PVT"])

x_lh_ms, y_lh_ms, min_lh_ms, max_lh_ms = compute_ecdf(lh, ["MS", "NDB"])
x_nac_ms, y_nac_ms, min_nac_ms, max_nac_ms = compute_ecdf(nac, ["MS", "NDB"])
x_pfc_ms, y_pfc_ms, min_pfc_ms, max_pfc_ms = compute_ecdf(pfc, ["MS", "NDB"])
```
