```python
#Load the .csv file for Lateral Hippocampus data
df_lh = pd.read_csv("LH_TRIO_cellcounts.csv")
df_lh.head(5)
```

```python
df_lh.columns
```

```python
figure1 = px.scatter_3d(df_lh, x = "AP", y = "ML", z = "DV", color = "color", opacity = 0.5)
    
figure1.update_layout(scene = dict(
                      xaxis=dict(
                          title=dict(
                              text='anterior-posterior'
                          )
                      ),
                      yaxis=dict(
                          title=dict(
                              text='medial-lateral'
                          )
                      ),
                      zaxis=dict(
                          title=dict(
                              text='dorsal-ventral'
                          )
                      ),
                    ),
                    width=1200, height=800, autosize=False)
figure1.update_traces(marker_size=3)

figure1.show()
```
