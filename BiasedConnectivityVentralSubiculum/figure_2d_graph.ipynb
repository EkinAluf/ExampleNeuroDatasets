```python
# import packages

import pandas as pd
import numpy as np
import pdb
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib as mpl
```

```python
# set colors for LH, PFC, and NAc

colors = sns.color_palette('muted', 10)

PFC_color = "#FFDB6E"
NAc_color = "#BF43CC"
LH_color = "#87FFE1"
```

```python
# Load the .csv files

lh = pd.read_csv("LH_TRIO_cellcounts.csv")
nac = pd.read_csv("NAc_TRIO_cellcounts.csv")
pfc = pd.read_csv("PFC_TRIO_cellcounts.csv")
```

```python
# isolating the cells where the input is coming from RE

lh_re = lh[lh["acronym"] == "RE"]
nac_re = nac[nac["acronym"] == "RE"]
pfc_re = pfc[pfc["acronym"] == "RE"]

# extracting AP values

ap_lh_re = lh_re["AP"]
ap_nac_re = nac_re["AP"]
ap_pfc_re = pfc_re["AP"]

min_ap_re = np.min([ap_lh_re.min(), ap_nac_re.min(), ap_pfc_re.min()])
max_ap_re = np.max([ap_lh_re.max(), ap_nac_re.max(), ap_pfc_re.max()])

# sorting AP values in increasing order

x_lh_re = np.sort(ap_lh_re)
x_nac_re = np.sort(ap_nac_re)
x_pfc_re = np.sort(ap_pfc_re)

# calculating the cumulative cell proportion

y_lh_re = np.arange(1, len(x_lh_re) + 1) / len(x_lh_re)
y_nac_re = np.arange(1, len(x_nac_re) + 1) / len(x_nac_re)
y_pfc_re = np.arange(1, len(x_pfc_re) + 1) / len(x_pfc_re)




# isolating the cells where the input is coming from PA

lh_pa = lh[lh["acronym"] == "PA"]
nac_pa = nac[nac["acronym"] == "PA"]
pfc_pa = pfc[pfc["acronym"] == "PA"]

# extracting AP values

ap_lh_pa = lh_pa["AP"]
ap_nac_pa = nac_pa["AP"]
ap_pfc_pa = pfc_pa["AP"]

min_ap_pa = np.min([ap_lh_pa.min(), ap_nac_pa.min(), ap_pfc_pa.min()])
max_ap_pa = np.max([ap_lh_pa.max(), ap_nac_pa.max(), ap_pfc_pa.max()])

# sorting AP values in increasing order

x_lh_pa = np.sort(ap_lh_pa)
x_nac_pa = np.sort(ap_nac_pa)
x_pfc_pa = np.sort(ap_pfc_pa)

# calculating the cumulative cell proportion

y_lh_pa = np.arange(1, len(x_lh_pa) + 1) / len(x_lh_pa)
y_nac_pa = np.arange(1, len(x_nac_pa) + 1) / len(x_nac_pa)
y_pfc_pa = np.arange(1, len(x_pfc_pa) + 1) / len(x_pfc_pa)




# isolating the cells where the input is coming from MPO

lh_mpo = lh[lh["acronym"] == "MPO"]
nac_mpo = nac[nac["acronym"] == "MPO"]
pfc_mpo = pfc[pfc["acronym"] == "MPO"]

# extracting AP values

ap_lh_mpo = lh_mpo["AP"]
ap_nac_mpo = nac_mpo["AP"]
ap_pfc_mpo = pfc_mpo["AP"]

min_ap_mpo = np.min([ap_lh_mpo.min(), ap_nac_mpo.min(), ap_pfc_mpo.min()])
max_ap_mpo = np.max([ap_lh_mpo.max(), ap_nac_mpo.max(), ap_pfc_mpo.max()])

# sorting AP values in increasing order

x_lh_mpo = np.sort(ap_lh_mpo)
x_nac_mpo = np.sort(ap_nac_mpo)
x_pfc_mpo = np.sort(ap_pfc_mpo)

# calculating the cumulative cell proportion

y_lh_mpo = np.arange(1, len(x_lh_mpo) + 1) / len(x_lh_mpo)
y_nac_mpo = np.arange(1, len(x_nac_mpo) + 1) / len(x_nac_mpo)
y_pfc_mpo = np.arange(1, len(x_pfc_mpo) + 1) / len(x_pfc_mpo)





# isolating the cells where the input is coming from PVT

lh_pvt = lh[lh["acronym"] == "PVT"]
nac_pvt = nac[nac["acronym"] == "PVT"]
pfc_pvt = pfc[pfc["acronym"] == "PVT"]

# extracting AP values

ap_lh_pvt = lh_pvt["AP"]
ap_nac_pvt = nac_pvt["AP"]
ap_pfc_pvt = pfc_pvt["AP"]

min_ap_pvt = np.min([ap_lh_pvt.min(), ap_nac_pvt.min(), ap_pfc_pvt.min()])
max_ap_pvt = np.max([ap_lh_pvt.max(), ap_nac_pvt.max(), ap_pfc_pvt.max()])

# sorting AP values in increasing order

x_lh_pvt = np.sort(ap_lh_pvt)
x_nac_pvt = np.sort(ap_nac_pvt)
x_pfc_pvt = np.sort(ap_pfc_pvt)

# calculating the cumulative cell proportion

y_lh_pvt = np.arange(1, len(x_lh_pvt) + 1) / len(x_lh_pvt)
y_nac_pvt = np.arange(1, len(x_nac_pvt) + 1) / len(x_nac_pvt)
y_pfc_pvt = np.arange(1, len(x_pfc_pvt) + 1) / len(x_pfc_pvt)



# isolating the cells where the input is coming from MS and NDB

lh_ms_ndb = lh[lh["acronym"].isin(["MS", "NDB"])]
nac_ms_ndb = nac[nac["acronym"].isin(["MS", "NDB"])]
pfc_ms_ndb = pfc[pfc["acronym"].isin(["MS", "NDB"])]

# extracting AP values

ap_lh_ms_ndb = lh_ms_ndb["AP"]
ap_nac_ms_ndb = nac_ms_ndb["AP"]
ap_pfc_ms_ndb = pfc_ms_ndb["AP"]

min_ap_ms_ndb = np.min([ap_lh_ms_ndb.min(), ap_nac_ms_ndb.min(), ap_pfc_ms_ndb.min()])
max_ap_ms_ndb = np.max([ap_lh_ms_ndb.max(), ap_nac_ms_ndb.max(), ap_pfc_ms_ndb.max()])

# sorting AP values in increasing order

x_lh_ms_ndb = np.sort(ap_lh_ms_ndb)
x_nac_ms_ndb = np.sort(ap_nac_ms_ndb)
x_pfc_ms_ndb = np.sort(ap_pfc_ms_ndb)

# calculating the cumulative cell proportion

y_lh_ms_ndb = np.arange(1, len(x_lh_ms_ndb) + 1) / len(x_lh_ms_ndb)
y_nac_ms_ndb = np.arange(1, len(x_nac_ms_ndb) + 1) / len(x_nac_ms_ndb)
y_pfc_ms_ndb = np.arange(1, len(x_pfc_ms_ndb) + 1) / len(x_pfc_ms_ndb)
```

```python
# make 5 subplots

global_min_ap = np.min([min_ap_re, min_ap_pa, min_ap_mpo, min_ap_pvt, min_ap_ms_ndb])
global_max_ap = np.max([max_ap_re, max_ap_pa, max_ap_mpo, max_ap_pvt, max_ap_ms_ndb])

fig = plt.figure(figsize=(12, 8))
spec = mpl.gridspec.GridSpec(ncols=6, nrows=2, figure=fig)

PA_fig   = fig.add_subplot(spec[0, 0:2])
MPO_fig  = fig.add_subplot(spec[0, 2:4])
RE_fig   = fig.add_subplot(spec[0, 4:6])
PVT_fig  = fig.add_subplot(spec[1, 1:3])
MS_NDB_fig = fig.add_subplot(spec[1, 3:5])

# plot RE

RE_fig.plot(x_lh_re, y_lh_re, color=LH_color, linewidth=3, label="LH")
RE_fig.plot(x_pfc_re, y_pfc_re, color=PFC_color, linewidth=3, label="PFC")
RE_fig.plot(x_nac_re, y_nac_re, color=NAc_color, linewidth=3, label="NAc")

RE_fig.set_xlabel("AP position")
RE_fig.set_ylabel("Cell proportion")
RE_fig.set_xlim(global_min_ap, global_max_ap)
RE_fig.set_ylim(0, 1.02)
RE_fig.legend(frameon=False)
RE_fig.set_title("RE")

# plot PA

PA_fig.plot(x_lh_pa, y_lh_pa, color=LH_color, linewidth=3, label="LH")
PA_fig.plot(x_pfc_pa, y_pfc_pa, color=PFC_color, linewidth=3, label="PFC")
PA_fig.plot(x_nac_pa, y_nac_pa, color=NAc_color, linewidth=3, label="NAc")

PA_fig.set_xlabel("AP position")
PA_fig.set_ylabel("Cell proportion")
PA_fig.set_xlim(global_min_ap, global_max_ap)
PA_fig.set_ylim(0, 1.02)
PA_fig.legend(frameon=False)
PA_fig.set_title("PA")

# plot MPO

MPO_fig.plot(x_lh_mpo, y_lh_mpo, color=LH_color, linewidth=3, label="LH")
MPO_fig.plot(x_pfc_mpo, y_pfc_mpo, color=PFC_color, linewidth=3, label="PFC")
MPO_fig.plot(x_nac_mpo, y_nac_mpo, color=NAc_color, linewidth=3, label="NAc")

MPO_fig.set_xlabel("AP position")
MPO_fig.set_ylabel("Cell proportion")
MPO_fig.set_xlim(global_min_ap, global_max_ap)
MPO_fig.set_ylim(0, 1.02)
MPO_fig.legend(frameon=False)
MPO_fig.set_title("MPO")

# plot PVT

PVT_fig.plot(x_lh_pvt, y_lh_pvt, color=LH_color, linewidth=3, label="LH")
PVT_fig.plot(x_pfc_pvt, y_pfc_pvt, color=PFC_color, linewidth=3, label="PFC")
PVT_fig.plot(x_nac_pvt, y_nac_pvt, color=NAc_color, linewidth=3, label="NAc")

PVT_fig.set_xlabel("AP position")
PVT_fig.set_ylabel("Cell proportion")
PVT_fig.set_xlim(global_min_ap, global_max_ap)
PVT_fig.set_ylim(0, 1.02)
PVT_fig.legend(frameon=False)
PVT_fig.set_title("PVT")

# plot MS + NDB

MS_NDB_fig.plot(x_lh_ms_ndb, y_lh_ms_ndb, color=LH_color, linewidth=3, label="LH")
MS_NDB_fig.plot(x_pfc_ms_ndb, y_pfc_ms_ndb, color=PFC_color, linewidth=3, label="PFC")
MS_NDB_fig.plot(x_nac_ms_ndb, y_nac_ms_ndb, color=NAc_color, linewidth=3, label="NAc")

MS_NDB_fig.set_xlabel("AP position")
MS_NDB_fig.set_ylabel("Cell proportion")
MS_NDB_fig.set_xlim(global_min_ap, global_max_ap)
MS_NDB_fig.set_ylim(0, 1.02)
MS_NDB_fig.legend(frameon=False)
MS_NDB_fig.set_title("MS/NDB")

plt.show()
```
